## 1 Пароль
Напишіть програму (функцію) для генерування випадкового пароля з такими обмеженнями:
- пароль складається з рівно $10$ символів
- пароль містить хоча б одну маленьку латинську літеру [a-z]
- пароль містить хоча б одну велику латинську літеру [A-Z]
- пароль містить хоча б одну цифру [0-9]
- пароль містить хоча б один спеціальний символ [~!@#$%^&*()-_+=\/{}[].,?<>:;]

```python
# Місце для Вашого коду
import numpy as np

def password():
    low_let = [chr(i) for i in range(ord('a'), ord('z'))]
    high_let = [chr(i) for i in range(ord('A'), ord('Z'))]
    symb = ["~", "!", "@", "#", "$", "%", "^", "&", "*", "(", ")", "-", "_", "+", "=", "/", "{", "}", "[", "]", ".", ",", "?", "<", ">", ":", ";"]

    # спочатку додаємо до пароля одну маленьку літеру, одну велику, одну цифру і один символ
    p = [-1 for i in range(10)]
    k, l, h, s, n = 0, 0, 0, 0, 0
    while k != 4:
        index = np.random.randint(10)
        if p[index] == -1 and l == 0:
            p[index] = np.random.choice(low_let, 1, True)[0]
            k += 1
            l = 1
        elif p[index] == -1 and h == 0:
            p[index] = np.random.choice(high_let, 1, True)[0]
            k += 1
            h = 1
        elif p[index] == -1 and s == 0:
            p[index] = np.random.choice(symb, 1, True)[0]
            k += 1
            s = 1
        elif p[index] == -1 and n == 0:
            p[index] = str(np.random.randint(10))
            k += 1
            n = 1
    
    # тепер на всі вільні місця в масиві ставимо випадкові символи, числа чи літери        
    for i in range(10):
        var = np.random.randint(4)
        if var == 1 and p[i] == -1:
            p[i] = np.random.choice(low_let, 1, True)[0]
        elif var == 2 and p[i] == -1:
            p[i] = np.random.choice(high_let, 1, True)[0]
        elif var == 3 and p[i] == -1:
            p[i] = np.random.choice(symb, 1, True)[0]
        elif var == 0 and p[i] == -1:
            p[i] = str(np.random.randint(10))
    
    # пертворення масиву символів у рядок
    pas = ""
    for i in p:
        pas += i
    return pas

    
print(password())    
```

```python
# Місце для Вашого коду
import numpy as np

def func(k, n):
    broski = np.random.choice(["H", "T"], n)
    PARuGerbiv = 0
    i = 0
    while i != len(broski) - 1:
        kk = 0
        if broski[i] == "H" and i != len(broski) - 1:
            kk += 1
            for j in range(i + 1, len(broski)):
                if broski[j] == "H":
                    kk += 1
                    if kk == k:
                        PARuGerbiv += 1
                        kk -= 1
                    if j == len(broski) - 1:
                        i += 1
                else:
                    i = j
                    break
        else:
            i += 1
    p = PARuGerbiv/(n - k + 1)
    return PARuGerbiv

N = [10, 50, 100, 500, 1000, 5000, 10000]
for n in N:
    k1 = func(2, n)
    k2 = func(3, n)
    print('n = {}, 2 герби підряд  :  Кількість випадінь =  {}, Йморність = {}'.format(n, k1, k1/(n - 2 + 1)))
    print('n = {}, 3 герби підряд  :  Кількість випадінь =  {}, Йморність = {}'.format(n, k2, k2/(n - 3 + 1)))
```

## 3 Найбільше число
Реалізуйте функцію підрахунку точної (теоретичної) імовірності того, що максимальне число після $n$ кидків звичайного грального кубика дорівнює $x\_max$, для таких параметрів:
- $n = 3, x\_max = 6$
- $n = 2, x\_max = 3$
- $n = 3, x\_max = 5$

Тобто, якщо $X_i, i = 1, \ldots, n$, - число на $i$-му кидку кубика, а $X = \max\limits_{1 \leq i \leq n} X_i$, то потрібно знайти ймовірність $\textbf{P} \left( X = x\_max \right)$.

Підрахуйте також емпіричні ймовірності відповідних подій у випадку $10000$ симуляцій.

**Зауваження.** Якщо введені параметри не мають сенсу, то повертати $0$.
Тобто, якщо вводиться $x\_max < 1$ або $x\_max > 6$.

```python
# Місце для Вашого коду
from itertools import product
import numpy as np

def func(n, x_max):
    if x_max < 1 or x_max > 6 or n < 1:
        return 0
    
    # теоретична йморність
    p = 0
    p_all = 0
    for x in product(range(1,7), repeat = n):
        if max(x) == x_max:
            p += 1
        p_all += 1
    print('n = {}, x_max = {}, Теоретична йморність : {}'.format(n, x_max, p/p_all))
    
    # емпірична йморність
    k = 0
    for i in range(10000):
        broski_emp = np.random.choice([1, 2, 3, 4, 5, 6], n)
        if max(broski_emp) == x_max:
            k += 1
    emp = k/10000
    print('Емпірична йморність: {}'.format(emp))
    
func(3, 6)
func(2, 3)
func(3, 5)
```

## 4 Сума чисел
Реалізуйте функцію підрахунку точної (теоретичної) імовірності того, що сума чисел після $n$ кидків звичайного грального кубика дорівнює $s$, для таких параметрів:
- $n = 3, s = 13$
- $n = 2, s = 5$
- $n = 3, s = 7$

Тобто, якщо $X_i, i = 1, \ldots, n$, - число на $i$-му кидку кубика, а $X = \sum\limits_{i = 1}^{n} X_i$, то потрібно знайти ймовірність $\textbf{P} \left( X = s \right)$.

Підрахуйте також емпіричні ймовірності відповідних подій у випадку $10000$ симуляцій.

**Зауваження.** Якщо введені параметри не мають сенсу, то повертати $0$.
Наприклад, якщо при $n = 3$ вводиться $s < 3$ або $s > 18$.

```python
# Місце для Вашого коду
from itertools import product
import numpy as np

def func(n, s):
    if n < 1 or s < n or s > 6*n:
        return 0
    
    # теоретична йморність
    p = 0
    p_all = 0
    for x in product(range(1,7), repeat = n):
        if sum(x) == s:
            p += 1
        p_all += 1
    print('n = {}, s = {}, Теоретична йморність : {}'.format(n, s, p/p_all))
    
    # емпірична йморність
    k = 0
    for i in range(10000):
        broski_emp = np.random.choice([1, 2, 3, 4, 5, 6], n)
        if sum(broski_emp) == s:
            k += 1
    emp = k/10000
    print('Емпірична йморність: {}'.format(emp))
    
func(3, 13)
func(2, 5)
func(3, 7)
```
